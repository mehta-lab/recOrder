```python
%%time
data = []
for path in data_paths_sort:
    obj = tiff.TiffFile(path)
    imgs = obj.asarray(key=range(0,len(obj.pages)))
    for img in imgs:
        data.append(img)
        
#     print(np.shape(data))
```
